```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Define constants
num_records = 1000
provinces = ['Gauteng', 'Western Cape', 'KwaZulu-Natal', 'Eastern Cape',
             'Free State', 'Limpopo', 'Mpumalanga', 'North West', 'Northern Cape']

industries = [
    'Retail & Wholesale', 'Food & Beverage', 'Agriculture', 'Manufacturing',
    'Construction', 'Technology', 'Financial Services', 'Healthcare',
    'Transportation & Logistics', 'Tourism & Hospitality', 'Education'
]

application_statuses = ['Approved', 'Rejected', 'Pending']
status_weights = [0.45, 0.40, 0.15]  # 45% approved, 40% rejected, 15% pending

# Generate business IDs
business_ids = [f'B{str(i).zfill(6)}' for i in range(1, num_records + 1)]

# Generate application dates within last 2 years
start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 10, 15)
days_range = (end_date - start_date).days
application_dates = [start_date + timedelta(days=random.randint(0, days_range)) for _ in range(num_records)]

# Generate requested amounts (R50,000 to R5,000,000)
requested_amounts = np.random.lognormal(mean=13, sigma=1, size=num_records).astype(int)
requested_amounts = np.clip(requested_amounts, 50000, 5000000)

# Generate application statuses
application_statuses = np.random.choice(application_statuses, size=num_records, p=status_weights)

# Generate approved amounts (0 for rejected, some percentage of requested for approved)
approved_amounts = []
for i in range(num_records):
    if application_statuses[i] == 'Approved':
        approved_amounts.append(int(requested_amounts[i] * np.random.uniform(0.5, 1.0)))
    elif application_statuses[i] == 'Pending':
        approved_amounts.append(0)
    else:  # Rejected
        approved_amounts.append(0)

# Generate loan purposes
loan_purposes = np.random.choice([
    'Working Capital', 'Equipment Purchase', 'Expansion', 'Inventory',
    'Debt Refinancing', 'Property Acquisition', 'Startup Funding'
], size=num_records)

# Generate loan terms (months)
loan_terms = np.random.choice([12, 24, 36, 48, 60, 72], size=num_records)

# Generate interest rates (8% to 25%)
interest_rates = np.random.uniform(8, 25, size=num_records).round(2)

# Create the dataframe
loan_apps = pd.DataFrame({
    'BusinessID': business_ids,
    'ApplicationDate': application_dates,
    'RequestedAmount': requested_amounts,
    'ApplicationStatus': application_statuses,
    'ApprovedAmount': approved_amounts,
    'LoanPurpose': loan_purposes,
    'LoanTerm': loan_terms,
    'InterestRate': interest_rates,
    'ProjectedJobs': np.random.randint(1, 50, size=num_records)
})

# Some records will have missing values for loan amount to simulate data quality issues
mask = np.random.choice([True, False], size=num_records, p=[0.05, 0.95])
loan_apps.loc[mask, 'LoanAmount'] = np.nan

# Save to CSV
loan_apps.to_csv('sme_loan_applications.csv', index=False)
print("Generated SME loan applications dataset with", num_records, "records")
print(loan_apps.head())
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Define constants
num_records = 1000  # Same as loan applications to ensure complete matching
provinces = ['Gauteng', 'Western Cape', 'KwaZulu-Natal', 'Eastern Cape',
             'Free State', 'Limpopo', 'Mpumalanga', 'North West', 'Northern Cape']

industries = [
    'Retail & Wholesale', 'Food & Beverage', 'Agriculture', 'Manufacturing',
    'Construction', 'Technology', 'Financial Services', 'Healthcare',
    'Transportation & Logistics', 'Tourism & Hospitality', 'Education'
]

# Use the same business IDs as in loan applications
business_ids = [f'B{str(i).zfill(6)}' for i in range(1, num_records + 1)]

# Generate registration dates (1-20 years ago)
current_date = datetime(2024, 10, 15)
registration_dates = []
for _ in range(num_records):
    years_ago = random.uniform(1, 20)
    days_ago = int(years_ago * 365)
    registration_dates.append(current_date - timedelta(days=days_ago))

# Calculate business age in years
business_ages = [(current_date - date).days / 365 for date in registration_dates]

# Generate company types
company_types = np.random.choice(['Pty Ltd', 'Sole Proprietor', 'Close Corporation', 'Partnership'], size=num_records, p=[0.5, 0.3, 0.15, 0.05])

# Generate employee counts (1 to 200 employees)
employee_counts = np.random.lognormal(mean=2.5, sigma=1.2, size=num_records).astype(int)
employee_counts = np.clip(employee_counts, 1, 200)

# Generate annual turnover (R100,000 to R50,000,000)
annual_turnover = np.random.lognormal(mean=15, sigma=1.5, size=num_records).astype(int)
annual_turnover = np.clip(annual_turnover, 100000, 50000000)

# Generate profit margins (-10% to 35%)
profit_margins = np.random.normal(loc=0.12, scale=0.08, size=num_records).round(3)
profit_margins = np.clip(profit_margins, -0.1, 0.35)

# Generate asset values (R50,000 to R20,000,000)
asset_values = np.random.lognormal(mean=14, sigma=1.3, size=num_records).astype(int)
asset_values = np.clip(asset_values, 50000, 20000000)

# Create the dataframe
business_data = pd.DataFrame({
    'BusinessID': business_ids,
    'RegistrationDate': registration_dates,
    'BusinessAge': business_ages,
    'CompanyType': company_types,
    'Industry': np.random.choice(industries, size=num_records),
    'Province': np.random.choice(provinces, size=num_records),
    'EmployeeCount': employee_counts,
    'AnnualTurnover': annual_turnover,
    'ProfitMargin': profit_margins,
    'AssetValue': asset_values
})

# Save to CSV
business_data.to_csv('sa_business_registry.csv', index=False)
print("Generated SA business registry dataset with", num_records, "records")
print(business_data.head())
```

```python
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Define provinces and sectors
provinces = ['Gauteng', 'Western Cape', 'KwaZulu-Natal', 'Eastern Cape',
             'Free State', 'Limpopo', 'Mpumalanga', 'North West', 'Northern Cape']

# Define sector categories (mapped from industries mentioned in the script)
sector_categories = ['Trade', 'Manufacturing', 'Agriculture', 'Construction',
                     'Technology', 'Financial Services', 'Healthcare', 'Transport',
                     'Tourism', 'Education']

# Create all combinations of province and sector
data = []
for province in provinces:
    for sector in sector_categories:
        data.append({'Province': province, 'SectorCategory': sector})

# Convert to DataFrame
economic_data = pd.DataFrame(data)

# Add province-level data
province_gdp = {
    'Gauteng': 1600000000000,  # R1.6 trillion
    'Western Cape': 800000000000,  # R800 billion
    'KwaZulu-Natal': 650000000000,  # R650 billion
    'Eastern Cape': 280000000000,  # R280 billion
    'Free State': 180000000000,  # R180 billion
    'Limpopo': 220000000000,  # R220 billion
    'Mpumalanga': 240000000000,  # R240 billion
    'North West': 170000000000,  # R170 billion
    'Northern Cape': 120000000000,  # R120 billion
}

province_population = {
    'Gauteng': 15000000,
    'Western Cape': 7000000,
    'KwaZulu-Natal': 11000000,
    'Eastern Cape': 6500000,
    'Free State': 2800000,
    'Limpopo': 5700000,
    'Mpumalanga': 4500000,
    'North West': 4000000,
    'Northern Cape': 1200000,
}

province_unemployment = {
    'Gauteng': 0.35,
    'Western Cape': 0.27,
    'KwaZulu-Natal': 0.37,
    'Eastern Cape': 0.43,
    'Free State': 0.36,
    'Limpopo': 0.41,
    'Mpumalanga': 0.38,
    'North West': 0.40,
    'Northern Cape': 0.39,
}

# Create provincial GDP per capita
province_gdp_per_capita = {province: province_gdp[province]/province_population[province] for province in provinces}

# Add these to the dataframe
economic_data['ProvincialGDP'] = economic_data['Province'].map(province_gdp)
economic_data['Population'] = economic_data['Province'].map(province_population)
economic_data['UnemploymentRate'] = economic_data['Province'].map(province_unemployment)
economic_data['GDPPerCapita'] = economic_data['Province'].map(province_gdp_per_capita)

# Create sector contributions to GDP for each province
# First, define baseline sector contributions
sector_contributions = {
    'Trade': 0.18,
    'Manufacturing': 0.15,
    'Agriculture': 0.03,
    'Construction': 0.04,
    'Technology': 0.08,
    'Financial Services': 0.22,
    'Healthcare': 0.07,
    'Transport': 0.10,
    'Tourism': 0.08,
    'Education': 0.05
}

# Apply some variation by province
economic_data['SectorGDPContribution'] = 0.0
for i, row in economic_data.iterrows():
    province = row['Province']
    sector = row['SectorCategory']

    # Base contribution
    base_contribution = sector_contributions[sector]

    # Add province-specific variation
    # For example, tourism is stronger in Western Cape, manufacturing stronger in Gauteng, etc.
    province_factor = 1.0  # Default is no change

    if province == 'Western Cape' and sector == 'Tourism':
        province_factor = 1.4
    elif province == 'Gauteng' and sector == 'Financial Services':
        province_factor = 1.3
    elif province == 'KwaZulu-Natal' and sector == 'Manufacturing':
        province_factor = 1.2
    elif province == 'Eastern Cape' and sector == 'Agriculture':
        province_factor = 1.3
    elif province == 'Free State' and sector == 'Agriculture':
        province_factor = 1.5
    elif province == 'Limpopo' and sector == 'Mining':
        province_factor = 1.6
    elif province == 'Mpumalanga' and sector == 'Mining':
        province_factor = 1.5
    elif province == 'Northern Cape' and sector == 'Mining':
        province_factor = 1.7

    # Apply some random variation (±15%)
    random_factor = np.random.uniform(0.85, 1.15)

    # Calculate the final contribution
    contribution = base_contribution * province_factor * random_factor

    economic_data.at[i, 'SectorGDPContribution'] = contribution

# Normalize sector contributions within each province to sum to 1
for province in provinces:
    province_mask = economic_data['Province'] == province
    total_contribution = economic_data.loc[province_mask, 'SectorGDPContribution'].sum()
    economic_data.loc[province_mask, 'SectorGDPContribution'] = economic_data.loc[province_mask, 'SectorGDPContribution'] / total_contribution

# Add sector-specific growth rates
economic_data['SectorGrowthRate'] = 0.0
for i, row in economic_data.iterrows():
    sector = row['SectorCategory']

    # Base growth rates
    if sector == 'Technology':
        base_growth = 0.06  # 6% growth
    elif sector == 'Healthcare':
        base_growth = 0.045  # 4.5% growth
    elif sector == 'Financial Services':
        base_growth = 0.04  # 4% growth
    elif sector == 'Tourism':
        base_growth = 0.03  # 3% growth
    elif sector == 'Education':
        base_growth = 0.025  # 2.5% growth
    elif sector == 'Trade':
        base_growth = 0.02  # 2% growth
    elif sector == 'Manufacturing':
        base_growth = 0.015  # 1.5% growth
    elif sector == 'Transport':
        base_growth = 0.02  # 2% growth
    elif sector == 'Construction':
        base_growth = 0.01  # 1% growth
    elif sector == 'Agriculture':
        base_growth = 0.005  # 0.5% growth
    else:
        base_growth = 0.02  # Default 2% growth

    # Add some random variation (±2%)
    random_variation = np.random.uniform(-0.02, 0.02)

    economic_data.at[i, 'SectorGrowthRate'] = base_growth + random_variation

# Add sector job creation potential (jobs per million Rand invested)
economic_data['JobsPerMillionInvested'] = 0.0
for i, row in economic_data.iterrows():
    sector = row['SectorCategory']

    # Base job creation rates
    if sector == 'Agriculture':
        base_jobs = 25  # 25 jobs per million
    elif sector == 'Manufacturing':
        base_jobs = 18  # 18 jobs per million
    elif sector == 'Construction':
        base_jobs = 22  # 22 jobs per million
    elif sector == 'Tourism':
        base_jobs = 20  # 20 jobs per million
    elif sector == 'Trade':
        base_jobs = 15  # 15 jobs per million
    elif sector == 'Transport':
        base_jobs = 12  # 12 jobs per million
    elif sector == 'Education':
        base_jobs = 16  # 16 jobs per million
    elif sector == 'Healthcare':
        base_jobs = 14  # 14 jobs per million
    elif sector == 'Financial Services':
        base_jobs = 6  # 6 jobs per million
    elif sector == 'Technology':
        base_jobs = 8  # 8 jobs per million
    else:
        base_jobs = 10  # Default 10 jobs per million

    # Add some random variation (±20%)
    random_variation = np.random.uniform(0.8, 1.2)

    economic_data.at[i, 'JobsPerMillionInvested'] = base_jobs * random_variation

# Save to CSV
economic_data.to_csv('provincial_economic_indicators.csv', index=False)
print("Generated provincial economic indicators dataset with", len(economic_data), "records")
print(economic_data.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd

# Set visualization style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Load datasets
print("Loading datasets...")
loan_apps = pd.read_csv('sme_loan_applications.csv')
business_data = pd.read_csv('sa_business_registry.csv')
economic_data = pd.read_csv('provincial_economic_indicators.csv')

print("Dataset sizes:")
print(f"Loan applications: {loan_apps.shape}")
print(f"Business registry: {business_data.shape}")
print(f"Economic indicators: {economic_data.shape}")

# Merge datasets on business registration number
merged_df = loan_apps.merge(business_data, on='BusinessID', how='left')
print(f"Merged dataset size: {merged_df.shape}")

# Check for missing values
print("\nMissing values in merged dataset:")
print(merged_df.isnull().sum())

# Handle missing values in critical columns
merged_df['LoanAmount'] = merged_df['LoanAmount'].fillna(0)
merged_df['BusinessAge'] = merged_df['BusinessAge'].fillna(merged_df['BusinessAge'].median())

# Create new variables
merged_df['ApprovalRate'] = merged_df['ApplicationStatus'].apply(lambda x: 1 if x == 'Approved' else 0)
merged_df['FundingGap'] = merged_df['RequestedAmount'] - merged_df['ApprovedAmount']

# Create sector categories that align with economic data
sector_map = {
    'Retail & Wholesale': 'Trade',
    'Food & Beverage': 'Manufacturing',
    'Agriculture': 'Agriculture',
    'Manufacturing': 'Manufacturing',
    'Construction': 'Construction',
    'Technology': 'Technology',
    'Financial Services': 'Financial Services',
    'Healthcare': 'Healthcare',
    'Transportation & Logistics': 'Transport',
    'Tourism & Hospitality': 'Tourism',
    'Education': 'Education'
}
merged_df['SectorCategory'] = merged_df['Industry'].map(sector_map)

# Join with economic data
analysis_df = merged_df.merge(economic_data, on=['Province', 'SectorCategory'], how='left')
print(f"Final analysis dataset size: {analysis_df.shape}")

# Create a simple risk score proxy
merged_df['RiskScore'] = (
    merged_df['BusinessAge'] * 0.3 +
    merged_df['ProfitMargin'] * 0.4 +
    merged_df['AssetValue'] * 0.3
) / (merged_df['RequestedAmount'] / 100000)

# Create output directory for plots
import os
if not os.path.exists('plots'):
    os.makedirs('plots')

# CORE ANALYSIS: IDENTIFYING FUNDING GAPS
print("\nAnalyzing approval rates and funding gaps by sector...")

# Approval rates by sector
sector_approval = merged_df.groupby('SectorCategory')['ApprovalRate'].mean().sort_values()

plt.figure()
sector_approval.plot(kind='barh')
plt.title('Loan Approval Rates by Sector')
plt.xlabel('Approval Rate (%)')
plt.tight_layout()
plt.savefig('plots/sector_approval_rates.png')

# Calculate funding gap by sector
sector_gaps = merged_df.groupby('SectorCategory')['FundingGap'].sum().sort_values(ascending=False) / 1000000

plt.figure()
sector_gaps.plot(kind='bar')
plt.title('Total Funding Gap by Sector (Million Rand)')
plt.ylabel('Funding Gap (Million Rand)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plots/sector_funding_gaps.png')

# Normalize by sector GDP contribution
merged_df['NormalizedGap'] = merged_df['FundingGap'] / merged_df.groupby('SectorCategory')['FundingGap'].transform('sum')
norm_gaps = merged_df.groupby('SectorCategory')['NormalizedGap'].mean().sort_values(ascending=False)

plt.figure()
norm_gaps.plot(kind='bar')
plt.title('Funding Gap Relative to Economic Contribution')
plt.tight_layout()
plt.savefig('plots/normalized_funding_gaps.png')

# GEOGRAPHIC ANALYSIS
print("\nAnalyzing geographic distribution of funding gaps...")

# Geographic analysis
province_gaps = merged_df.groupby('Province')['FundingGap'].sum() / merged_df.groupby('Province')['RequestedAmount'].sum()

try:
    # Try to load the South Africa province map
    sa_map = gpd.read_file('south_africa_provinces.geojson')

    # Merge with funding gap data
    sa_map = sa_map.merge(pd.DataFrame(province_gaps).reset_index(), left_on='PROVINCE', right_on='Province')

    # Plot the map
    fig, ax = plt.subplots(1, 1)
    sa_map.plot(column='FundingGap', ax=ax, legend=True, cmap='OrRd')
    plt.title('Funding Gap by Province (% of Requested Amount)')
    plt.savefig('plots/province_funding_gaps_map.png')

except Exception as e:
    print(f"Warning: Could not create map visualization: {e}")

    # Create a bar chart instead
    plt.figure()
    province_gaps.sort_values().plot(kind='barh')
    plt.title('Funding Gap by Province (% of Requested Amount)')
    plt.tight_layout()
    plt.savefig('plots/province_funding_gaps_bar.png')

# RISK ANALYSIS
print("\nAnalyzing risk factors and approval rates...")

# Plot risk score vs approval probability
plt.figure()
sns.regplot(x='RiskScore', y='ApprovalRate', data=merged_df, logistic=True)
plt.title('Risk Score vs. Approval Probability')
plt.xlabel('Risk Score (higher = less risky)')
plt.ylabel('Approval Probability')
plt.savefig('plots/risk_vs_approval.png')

# Compare risk-adjusted approval rates by sector
merged_df['RiskBucket'] = pd.qcut(merged_df['RiskScore'], 5, labels=['Very High', 'High', 'Medium', 'Low', 'Very Low'])
risk_sector_approval = pd.pivot_table(
    merged_df,
    values='ApprovalRate',
    index='SectorCategory',
    columns='RiskBucket',
    aggfunc='mean'
)

plt.figure(figsize=(14, 8))
sns.heatmap(risk_sector_approval, annot=True, cmap='YlGnBu')
plt.title('Approval Rates by Sector and Risk Level')
plt.tight_layout()
plt.savefig('plots/risk_adjusted_approval_rates.png')

# SOLUTION EXPLORATION
print("\nExploring alternative lending models...")

# Simulate an alternative lending model that weights job creation potential
merged_df['JobsPerMillion'] = merged_df['ProjectedJobs'] / (merged_df['RequestedAmount'] / 1000000)
merged_df['AlternativeScore'] = (
    merged_df['RiskScore'] * 0.7 +
    merged_df['JobsPerMillion'].rank(pct=True) * 0.3
)

# Compare approval patterns
original_approvals = merged_df.groupby('SectorCategory')['ApprovalRate'].mean()
alt_scores = merged_df.groupby('SectorCategory')['AlternativeScore'].mean()
alt_approvals = alt_scores / alt_scores.max()  # Normalize to 0-1 range

comparison = pd.DataFrame({
    'Current': original_approvals,
    'Alternative': alt_approvals
})

plt.figure()
comparison.plot(kind='bar')
plt.title('Current vs. Alternative Lending Model (Sector Approval Rates)')
plt.ylabel('Approval Rate / Score')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plots/current_vs_alternative_model.png')

# Additional analysis: Funding gap by business size
print("\nAnalyzing funding gaps by business size...")

# Create business size categories
merged_df['BusinessSize'] = pd.cut(
    merged_df['EmployeeCount'],
    bins=[0, 5, 20, 50, 200],
    labels=['Micro (1-5)', 'Small (6-20)', 'Medium (21-50)', 'Large (51+)']
)

# Calculate funding gap by business size
size_gaps = merged_df.groupby('BusinessSize')['FundingGap'].sum() / merged_df.groupby('BusinessSize')['RequestedAmount'].sum()

plt.figure()
size_gaps
```
